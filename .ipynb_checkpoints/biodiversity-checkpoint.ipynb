```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# load up all data
park_data = pd.read_csv('observations.csv')
species_data = pd.read_csv('species_info.csv')
```

```python
# see details about data
print(park_data.info())
print()
print(species_data.info())
```

```python
# visualize the frequency of different categories of animals
f, ax = plt.subplots(figsize=(12,7))
sns.histplot(data=species_data, x='category')
plt.show()
```

```python
# visualize the frequency of conservation status for species
f, ax = plt.subplots(figsize=(12,7))
sns.histplot(data=species_data, x='conservation_status')
plt.show()
```

```python
# visualize the number of obervations for each park
f, ax = plt.subplots(figsize=(12,7))
sns.barplot(data=park_data, x='park_name', y='observations')
plt.show()
```

```python
# visualize the frequency of conservation status for species
f, ax = plt.subplots(figsize=(12,7))
sns.histplot(data=species_data, x='conservation_status', y='category', cbar=True)
plt.show()
```

<h1>Analysis</h1>
<p> From the graphs we can make a few conclusions.
    <ul>
        <li>The majority of observations are plant based; most likely due to the large variety of plants, and their stationary nature. While animals operate in large areas where they hunt and move.</li>
        <li>A large amount of species are 'Species of Concern' which means that there are concerns for their population, but currently are not in danger of going extinct.</li>
        <li>Each of the parks are consistent with the number of obervations that they make for each species which is good for consistency, but some parks make a larger amount of observations than others.</li>
        <li> Birds, Mammals, and Vascular Plants are the category of species that have the largest amount of species that are Species of Concern.</li>
        </ul>
